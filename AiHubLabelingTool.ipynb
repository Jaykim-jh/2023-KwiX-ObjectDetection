```python
import os
import numpy as np
import shutil
import cv2
import json
```

```python
def convert(size, box):
    #box -> x_min, y_min, x_max, y_max (AiHub 파일 내 레이블링 방식)
    #size -> width, height
    dw = 1./size[0] # 이미지 내 X축 비율 표시용
    dh = 1./size[1] # 이미지 내 Y축 비율 표시용
    x = (box[0] + box[2]) / 2.0
    y = (box[1] + box[3]) / 2.0
    w = box[2] - box [0]
    h = box[3] - box[1]
    x = x*dw
    y = y*dh
    w = w*dw
    h = h*dh
    return (str(x), str(y), str(w), str(h))
count = 0
```

```python
IMAGE_PATH = 'C:\Download/New_sample/Source_Data/'#변환할 데이터 원본 이미지 파일 경로
LABEL_PATH = 'C:\Download/New_sample/Label_Data/'#변환할 데이터 원본 레이블 파일 경로
NEW_IMAGE_PATH = 'C:\Download/New_sample/Source_YOLO/'#새로 저장할 데이터 이미지 파일 경로
NEW_LABEL_PATH = 'C:\Download/New_sample/Label_YOLO/'#새로 저장할 데이터 레이블 파일 경로

IMAGE = sorted(os.listdir(IMAGE_PATH))
LABEL = sorted(os.listdir(LABEL_PATH))
```
